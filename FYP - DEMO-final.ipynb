# DEMO

```python
print("FINAL YEAR PROJECT - SENTIMENT ANALYSIS AND BOOK RECOMMENDATION BY TEAM 15\n")
print("The team members are:\n Lokesh Kumar - 18BC0S27 \n Shivali Muthukumar - 18BCS068 \n Kritya Shree S - 18BCS093\n\n")
    

def show_articles_more_than(column=['total score', 'sentiment_score']):
    df = All_ratings.sort_values('total score', ascending=False)
    return df

def genre(column=['total score', 'sentiment_score']):
    try:
        n = int(input("\nEnter your desired genre: \n1.children \n2.comics \n3.young \n4.mystery \n"))
        if n == 1:
            x = Children_ratings
        if n == 2:
            x = Comics_ratings
        if n == 3:
            x = Young_ratings
        if n == 4:
            x = Mystery_ratings
        if n > 4 or n < 0:
            return 'Enter a Valid Choice'
        return x.sort_values('total score', ascending=False).head(10)
    except:
        return 'Enter a valid integer!'

def random():
    try:
        a = int(input("1.Book with just the name \n2.Full description about the Book\n"))
    
        if a == 1:
            n = int(input("\nEnter your desired genre: \n1.children \n2.comics \n3.young \n4.mystery \n5.ANY \n"))
            if n == 1:
                x = pd.DataFrame(Children_ratings)
            if n == 2:
                x = pd.DataFrame(Comics_ratings)
            if n == 3:
                x = pd.DataFrame(Young_ratings)
            if n == 4:
                x = pd.DataFrame(Mystery_ratings)
            if n == 5:
                x = pd.DataFrame(All_ratings)
            if n > 5 or n <1:
                return '\nEnter a Valid Choice'
            return x.sample()
        if a == 2:
            n = int(input("\nEnter your desired genre: \n1.children \n2.comics \n3.young \n4.mystery \n5.ANY \n"))
            if n == 1:
                x = pd.DataFrame(Children)
            if n == 2:
                x = pd.DataFrame(Comics)
            if n == 3:
                x = pd.DataFrame(Young)
            if n == 4:
                x = pd.DataFrame(Mystery)
            if n == 5:
                x = pd.DataFrame(All_Genres)
            if n > 5 or n <1:
                return '\nEnter a Valid Choice'
            return x.sample()
        if a > 1 or a < 2:
            return '\nEnter a Valid Choice'
    except:
        return 'Enter a valid integer!'
        
def similar_books():
    df=All_Genres
    userRatings = df.pivot_table(index=['user_id'],columns=['title'],values='sentiment_score')
    item_util_matrix = userRatings.copy()
    item_util_matrix = item_util_matrix.apply(lambda col : col.fillna(col.mean()), axis=0)
    item_corr_matrix = item_util_matrix.corr()
    book = input('Enter the Book name\n')
    try:
        book_corr = item_corr_matrix[book]
        book_corr = book_corr.sort_values(ascending=False)
        book_corr.dropna(inplace=True)
        similar_books = pd.DataFrame(data=book_corr.values, columns=['Correlation'], 
                                               index = book_corr.index)
        return similar_books.head(6)
    except:
        return 'Bookname not found!'
    
def main():
    while True:
        print("\nSelect one of the following: \n1. Best ranked books \n2. Most recommended books in each genre \n3. Similar Books \n4. Random Suggestion\n5. Exit\n ")
        try:
            choice = int(input())
            if choice == 1:
                z=int(input("\n1.To display the best Book \n2.To display Top 5 Books \n3.To display Top 10 Books\n"))
                if z == 1:
                    display(show_articles_more_than().head(1))
                if z == 2:
                    display(show_articles_more_than().head(5))
                if z == 3:
                    display(show_articles_more_than().head(10))
                if z > 3 or z < 1 :
                    display('Enter a Valid Option')
            if choice == 2:
                display(genre())
            if choice == 3:
                display(similar_books())
            if choice == 4:
                display(random())
            if choice == 5:
                break
            if choice > 5 or choice < 1 :
                display('Enter a Valid Choice')
        except:
            display('Enter a valid integer!')
    
if __name__ == "__main__":
    main()
```
